```python
with open('/home/xgl470/combined.csv', 'rb') as f:
    reader = csv.reader(f)
    your_list = list(reader)
```

```python
test_list=your_list[0:100]
```

```python
def half(quarter):
    value=quarter
    if quarter=='1':
        value='1'
    elif quarter=='2':
        value='1'
    elif quarter=='3':
        value='2'
    elif quarter=='4':
        value='2'
    else:
        value='3'
    return value
```

```python
columns.append('half')
for x in test_list[1:]:
    half_value=half(x[columns.index('qtr')])
    x.append(half_value)
```

```python
columns.append('drive_result')
for x in test_list[1:]:
    index=test_list.index(x)
    for i in range(1,50):
        if (index+i)<len(test_list):
            if x[columns.index('gameid')]==test_list[index+i][columns.index('gameid')]:
                current_off=x[columns.index('off')]
                current_def=x[columns.index('def')]
                next_off=test_list[index+i][columns.index('off')]
                next_def=test_list[index+i][columns.index('def')] 
                current_off_score=int(x[columns.index('offscore')])
                current_def_score=int(x[columns.index('defscore')])
                next_off_score=int(test_list[index+i][columns.index('offscore')])
                next_def_score=int(test_list[index+i][columns.index('defscore')])
                if current_off==next_off and (current_off_score<>next_off_score or current_def_score<>next_def_score) and len(x)==14:
                    x.append((next_off_score-current_off_score)-(next_def_score-current_def_score))
                elif current_off<>next_off and len(x)==14:
                    x.append((next_def_score-current_off_score)-(next_off_score-current_def_score))
                    break
```

```python
columns.append('time')
for x in test_list[1:]:
    index=test_list.index(x)
    time=x[columns.index('description')]
    if ')' not in time and ')' in test_list[index-1][columns.index('description')]:
        time=test_list[index-1][columns.index('description')]
    elif ')' not in time and ')' in test_list[index+1][columns.index('description')]:
        time=test_list[index+1][columns.index('description')]
    else:
        time='(15:00)'
    left=time.index(')')
    time=time[1:left]
    time=int(time[0:time.index(':')])*60+int(time[time.index(':')+1:])
    x.append(time)
```

```python
columns.append('play_type')
for x in test_list[1:]:
    desc=x[columns.index('description')]
    if 'pass' in desc.lower():
        play_type='Pass'
    elif 'no play' in desc.lower():
        play_type='No Play'
    elif 'field goal' in desc.lower():
        play_type='Field Goal'
    elif 'punt' in desc.lower():
        play_type='Punt'
    elif 'kick' in desc.lower():
        play_type='Kickoff'
    else:
        play_type='Rush'
    x.append(play_type)
```
